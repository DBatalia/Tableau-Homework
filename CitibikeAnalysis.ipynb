```python
#df2018.starttime = df2018.starttime.apply(lambda d : dt.strptime(d, "%Y-%m-%d %H:%M:%S.%f"))
#df2018.head(50)
from datetime import datetime
def dateConvert(d):

   year=int(d[:4])

   month=int(d[5:7])

   day=int(d[8:10])

   hour=int(d[11:13])

   minute=int(d[14:16])

   seconds=int(d[17:19])

   a = datetime(year, month, day, hour, minute, seconds)

   return a.strftime("%m/%d/%Y %H:%M:%S")

df2017['starttime']=df2017['starttime'].apply(lambda d: dateConvert(d))

df2017['stoptime']=df2017['stoptime'].apply(lambda d: dateConvert(d))

df2017.head()
```

```python
dftotal=pd.concat([df2015, df2016,df2017])
dftotal.head()
```

```python
dftotal.to_csv("Citi.csv",index=False)
```
